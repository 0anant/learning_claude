```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
def add_user_messages(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages, system=None, temperature=1.0):
    params= {
        "model": model,
        "max_tokens":  100,
        "messages": messages,
        "temperature": temperature
    }
    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
messages = []

system = """You are a patience math tutor. you have to guide the student properly without giving answers directly"""

add_user_message = messages(messages, "How do i solve 5x+3=2 for x?")
answer = chat(messages, system = system, temperature=0.3)

answer
```
